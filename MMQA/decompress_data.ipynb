```python
# decompress all datasaets in a directory
import os
import gzip
import shutil

def decompress_gz_files(source_folder, destionation_folder):
    """
    Decompress all .gz files in the source_folder and save them to destination_folder.

    :param source_folder: The directory containing .gz files.
    :param destination_folder: The directory to save decompressed files.
    """
    # Ensure destination folder exists
    os.makedirs(destination_folder, exist_ok=True)

    # Iterate through files in the source folder
    for filename in os.listdir(source_folder):
        if filename.endswith(".gz"):
            source_file_path = os.path.join(source_folder, filename)
            output_file_path = os.path.join(
                destination_folder, filename[:-3]
            )  # Remove .gz extension

            with gzip.open(source_file_path, "rb") as f_in:
                with open(output_file_path, "wb") as f_out:
                    shutil.copyfileobj(f_in, f_out)

            print(f"Decompressed: {source_file_path} -> {output_file_path}")
source_folder = "dataset"
destination_folder = "jsonl_dataset"
decompress_gz_files(source_folder, destination_folder)
```
